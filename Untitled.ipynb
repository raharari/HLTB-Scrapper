```python
import pandas as pd
df = pd.read_csv('C:/Users/rharari/Desktop/Data Science/hltb1.csv', sep='|')
```

```python
df = df[0:10]
```

```python
#Lista de Generos
import itertools
unique_genres = df['Genres'].unique()
individual_genres = []
for genre in unique_genres:
    individual_genres.append(genre.split(','))

individual_genres = list(itertools.chain.from_iterable(individual_genres))
individual_genres = set(individual_genres)

individual_genres
```

```python
#Lista de Developers
import itertools
unique_devs = df['Developer'].unique()
individual_devs = []
for dev in unique_devs:
    individual_devs.append(dev.split(','))

individual_devs = list(itertools.chain.from_iterable(individual_devs))
individual_devs = set(individual_devs)

individual_devs
```

```python
#Lista de Publishers
import itertools
unique_pubs = df['Publishers'].unique()
individual_pubs = []
for pub in unique_pubs:
    individual_pubs.append(pub.split(','))

individual_pubs = list(itertools.chain.from_iterable(individual_pubs))
individual_pubs = set(individual_pubs)

individual_pubs
```

```python
import numpy as np
for genre in individual_genres:
    df['gen_'+genre] = np.nan
```

```python
# Como hago para iterar fila por file y sacar si es ese genero o no

for genre in individual_genres:
    for i in range(len(df)):
        if df['Genres'][i].find(genre) != -1:
            df['gen_'+genre] [i] = 1
        else:
            df['gen_'+genre] [i] = 0
```
